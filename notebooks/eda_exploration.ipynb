## 2. Data Quality Check


```python
# Generate quality report using Polars
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")

# Check for missing values
missing_values = {col: df[col].null_count() for col in df.columns if df[col].null_count() > 0}
print(f"Missing values: {len(missing_values)} columns with nulls")
if missing_values:
    for col, count in missing_values.items():
        print(f"  - {col}: {count} nulls")

# Check for duplicates
duplicate_count = df.height - df.unique().height
print(f"Duplicate rows: {duplicate_count}")
```
